# Most valuable players

```python
def value_to_int(df_value):
    value = float(df_value[1:-1])
    suffix = df_value[-1:]
    
    if suffix == 'M':
        value = value * 1000000
    elif suffix == 'K':
        value = value * 1000
        
    return value


df['Value_float'] = df['Value'].apply(value_to_int)
```

```python
df[['Value_float']]
```

```python
valuable_players = df.sort_values(by = ['Value_float'], ascending=False)
valuable_players[['Name', 'Age', 'Value']]
```
